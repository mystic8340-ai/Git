## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df.drop('Unnamed: 0',axis=1,inplace=True, errors='ignore')
Salary=df['Salary Estimate']
Salary=Salary.str.split('(').str[0]
Salary=Salary.str.replace('K','',regex=False)  # removing the unwanted values from salary column
Salary=Salary.str.replace('$','',regex=False)
Salary=Salary.str.replace('Employer Provided Salary:','',regex=False)
Salary=Salary.str.replace('Per Hour','',regex=False)
Salary=Salary.str.strip()
Salary=Salary[Salary !='-1']
df=df.loc[Salary.index]

#Extracting min and max vcalues from salary

df['min_salary']=Salary.apply(lambda x: int(x.split('-')[0].strip()))
df['max_salary']=Salary.apply(lambda x: int(x.split('-')[1].strip()))

# Finding the average salary
df['avg_salary']=(df['min_salary']+df['max_salary'])/2
print(df[['min_salary','max_salary','avg_salary']].head())

#Finding the companys age
df['Founded']=df['Founded'].replace(-1,np.nan)
df['Company_age']=2026-df['Founded']

#Finding the location or sates of employees
df['Job_state']=df['Location'].apply(lambda x:x.split(',')[-1].strip())

#Simplifing title
def simplify_title(title):
    title=title.lower()
    if 'data scientist'in title:
        return 'Data Scientist'
    elif 'data engineer'in title:
        return 'Data Engineer'
    elif 'data alanlyst'in title:
        return 'Data Analyst'
    elif 'machine learning'in title:
        return 'ML Engineer'
    else:
        return 'Other'

df['Job_Category']=df['Job Title'].apply(simplify_title)
````
